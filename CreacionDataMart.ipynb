```python
import pandas as pd
import numpy as np
```

```python
# EXTRACCIÓN DE DATOS DE LA CARPETA DATA/EDUCACION
```

```python
GastoEducacionPorcentajePBI = pd.read_csv("./data/Educacion/GastoEducacionPorcentajePBI.csv", delimiter=",")
contruccionInstitucionesPorcentajeGE = pd.read_csv("./data/Educacion/contruccionInstitucionesEducacion.csv", delimiter=",")
gastoInvestigacionPorcentajeGE = pd.read_csv("./data/Educacion/gastoInvestigacion.csv", delimiter=",")
subvencionesPorcentajeGE = pd.read_csv("./data/Educacion/SubvencionEscolar.csv", delimiter=",")
sueldoEmpleadosPorcentajeGE = pd.read_csv("./data/Educacion/sueldoMaestros.csv", delimiter=",")
```

```python
#PRIMER TRATAMIENTO DE DATOS 
#Se eliminan los valores no deseados y se nos quedamos solo con los valores objetivos
```

```python
GastoEducacionPorcentajePBI = GastoEducacionPorcentajePBI[['Reference Area','Time Period','Observation Value']]
GastoEducacionPorcentajePBI['pais_key'] = GastoEducacionPorcentajePBI['Reference Area']
GastoEducacionPorcentajePBI.drop('Reference Area', axis=1, inplace=True)
GastoEducacionPorcentajePBI['tiempo_key'] = GastoEducacionPorcentajePBI['Time Period']
GastoEducacionPorcentajePBI.drop('Time Period', axis=1, inplace=True)
GastoEducacionPorcentajePBI['porcentajeGastoEducacion'] = GastoEducacionPorcentajePBI['Observation Value']
GastoEducacionPorcentajePBI.drop('Observation Value', axis=1, inplace=True)
#-------------------------------------------------------------------------------------------------------------------
contruccionInstitucionesPorcentajeGE = contruccionInstitucionesPorcentajeGE[['Reference Area','Time Period','Observation Value']]
contruccionInstitucionesPorcentajeGE['pais_key'] = contruccionInstitucionesPorcentajeGE['Reference Area']
contruccionInstitucionesPorcentajeGE.drop('Reference Area', axis=1, inplace=True)
contruccionInstitucionesPorcentajeGE['tiempo_key'] = contruccionInstitucionesPorcentajeGE['Time Period']
contruccionInstitucionesPorcentajeGE.drop('Time Period', axis=1, inplace=True)
contruccionInstitucionesPorcentajeGE['porcentajeGastoEducacion'] = contruccionInstitucionesPorcentajeGE['Observation Value']
contruccionInstitucionesPorcentajeGE.drop('Observation Value', axis=1, inplace=True)
#-------------------------------------------------------------------------------------------------------------------
gastoInvestigacionPorcentajeGE = gastoInvestigacionPorcentajeGE[['Reference Area','Time Period','Observation Value']]
gastoInvestigacionPorcentajeGE['pais_key'] = gastoInvestigacionPorcentajeGE['Reference Area']
gastoInvestigacionPorcentajeGE.drop('Reference Area', axis=1, inplace=True)
gastoInvestigacionPorcentajeGE['tiempo_key'] = gastoInvestigacionPorcentajeGE['Time Period']
gastoInvestigacionPorcentajeGE.drop('Time Period', axis=1, inplace=True)
gastoInvestigacionPorcentajeGE['porcentajeGastoEducacion'] = gastoInvestigacionPorcentajeGE['Observation Value']
gastoInvestigacionPorcentajeGE.drop('Observation Value', axis=1, inplace=True)
#-------------------------------------------------------------------------------------------------------------------
subvencionesPorcentajeGE = subvencionesPorcentajeGE[['Reference Area','Time Period','Observation Value']]
subvencionesPorcentajeGE['pais_key'] = subvencionesPorcentajeGE['Reference Area']
subvencionesPorcentajeGE.drop('Reference Area', axis=1, inplace=True)
subvencionesPorcentajeGE['tiempo_key'] = subvencionesPorcentajeGE['Time Period']
subvencionesPorcentajeGE.drop('Time Period', axis=1, inplace=True)
subvencionesPorcentajeGE['porcentajeGastoEducacion'] = subvencionesPorcentajeGE['Observation Value']
subvencionesPorcentajeGE.drop('Observation Value', axis=1, inplace=True)
#-------------------------------------------------------------------------------------------------------------------
sueldoEmpleadosPorcentajeGE = sueldoEmpleadosPorcentajeGE[['Reference Area','Time Period','Observation Value']]
sueldoEmpleadosPorcentajeGE['pais_key'] = sueldoEmpleadosPorcentajeGE['Reference Area']
sueldoEmpleadosPorcentajeGE.drop('Reference Area', axis=1, inplace=True)
sueldoEmpleadosPorcentajeGE['tiempo_key'] = sueldoEmpleadosPorcentajeGE['Time Period']
sueldoEmpleadosPorcentajeGE.drop('Time Period', axis=1, inplace=True)
sueldoEmpleadosPorcentajeGE['porcentajeGastoEducacion'] = sueldoEmpleadosPorcentajeGE['Observation Value']
sueldoEmpleadosPorcentajeGE.drop('Observation Value', axis=1, inplace=True)

```

```python
#CREACIÓN DE LA DIMENSIÓN EDUCACIÓN
```

```python
educacion_dim = pd.DataFrame({
    'tipo_porcentaje_gasto_educacion': ["Porcentaje de gasto en educacion respecto al PBI",
                              "Gasto en contruccion de instituaciones respecto al gasto en educacion",
                              "Gasto en investigacion respecto al gasto en educacion",
                              "Gasto en subvencion escolar respecto al gasto en educacion",
                              "Gasto en sueldos a personal respecto al gasto en educacion"]
})
educacion_dim.index = educacion_dim.index+1
educacion_dim.index.name="educacion_key"
```

```python
#ASIGNACIÓN DE LAS CLAVES PRIMARIAS
```

```python
GastoEducacionPorcentajePBI=GastoEducacionPorcentajePBI.assign(educacion_key=1)
contruccionInstitucionesPorcentajeGE=contruccionInstitucionesPorcentajeGE.assign(educacion_key=2)
gastoInvestigacionPorcentajeGE=gastoInvestigacionPorcentajeGE.assign(educacion_key=3)
subvencionesPorcentajeGE=subvencionesPorcentajeGE.assign(educacion_key=4)
sueldoEmpleadosPorcentajeGE=sueldoEmpleadosPorcentajeGE.assign(educacion_key=5)
```

```python
#CREACIÓN DE LA DIMENSIÓN TIEMPO
```

```python
tiempo_dim = GastoEducacionPorcentajePBI.groupby(by= ["tiempo_key"]).mean()
tiempo_dim["agno"]= tiempo_dim.index
tiempo_dim= tiempo_dim.reset_index(drop=True)
tiempo_dim.index=tiempo_dim.index+1
tiempo_dim.index.name="tiempo_key"
tiempo_dim.drop('porcentajeGastoEducacion', axis=1, inplace=True)
tiempo_dim.drop('educacion_key', axis=1, inplace=True)
```

```python
#CREACIÓN DE LA DIMENSIÓN PAÍS
```

```python
pais_dim = GastoEducacionPorcentajePBI.groupby(by= ["pais_key"]).mean()
pais_dim["nombrePais"]= pais_dim.index
pais_dim= pais_dim.reset_index(drop=True)
pais_dim.index=pais_dim.index+1
pais_dim.index.name="pais_key"
pais_dim.drop('porcentajeGastoEducacion', axis=1, inplace=True)
pais_dim.drop('tiempo_key', axis=1, inplace=True)
pais_dim.drop('educacion_key', axis=1, inplace=True)
```

```python
#PRIMERA AGRUPACIÓN DE DATOS
#Añadimos los datos obtenidos de educación y los combinamos hasta obtener una sola tabla la que usaremos luego
```

```python
gasto_educacion = pd.DataFrame()
gasto_educacion = gasto_educacion.append(GastoEducacionPorcentajePBI)
gasto_educacion = gasto_educacion.append(contruccionInstitucionesPorcentajeGE)
gasto_educacion = gasto_educacion.append(gastoInvestigacionPorcentajeGE)
gasto_educacion = gasto_educacion.append(subvencionesPorcentajeGE)
gasto_educacion = gasto_educacion.append(sueldoEmpleadosPorcentajeGE)

gasto_educacion= gasto_educacion.reset_index(drop=True)
gasto_educacion.drop_duplicates(inplace=True)

agnos = pd.DataFrame()
agnos = agnos.append(tiempo_dim)
agnos['clave'] = agnos.index
clave_valor = agnos.to_numpy().transpose().tolist()
clave_valor
clave = clave_valor[1]
valor = clave_valor[0]

gasto_educacion["tiempo_key"] = gasto_educacion["tiempo_key"].replace(valor,clave)

```

```python
#MODIFICAMOS LOS VALORES POR SUS CLAVES RESPECTIVAS
```

```python
paises = pd.DataFrame()
paises = paises.append(pais_dim)

paises['clave'] = paises.index
p_clave_valor = paises.to_numpy().transpose().tolist()

p_clave = p_clave_valor[1]
p_valor = p_clave_valor[0]


gasto_educacion["pais_key"] = gasto_educacion["pais_key"].replace(p_valor,p_clave)
```

```python
#TRAEMOS LOS DATOS PARA SALUD
```

```python
gastoSaludPorcentajePBI = pd.read_csv("./data/Salud/porcentajeGastoSaludPBI.csv", delimiter=",")
gastoconstruccionInstitucionesMedicoGS = pd.read_csv("./data/Salud/contruccionInstituciones.csv", delimiter=",")
gastoMateriaMedicoGS = pd.read_csv("./data/Salud/materialMedico.csv", delimiter=",")
gastoSueldoPersonalMedicoGS = pd.read_csv("./data/Salud/gastoSueldoPersonalSalud.csv", delimiter=",")
```

```python
#CREAMOS SU DIMENSIÓN
```

```python
salud_dim = pd.DataFrame({
    'tipo_porcentaje_gasto_salud': ["Porcentaje de gasto en salud respecto al PBI",
                              "Gasto en contruccion de instituaciones respecto al gasto en salud",
                              "Gasto en compra material medico respecto al gasto en salud",
                              "Gasto en sueldos al personal respecto al gasto en salud"]
})
salud_dim.index = salud_dim.index+1
salud_dim.index.name="educacion_key"
```

```python
#ASIGNAMOS SUS CLAVES
```

```python
gastoSaludPorcentajePBI=gastoSaludPorcentajePBI.assign(salud_key=1)
gastoconstruccionInstitucionesMedicoGS=gastoconstruccionInstitucionesMedicoGS.assign(salud_key=2)
gastoMateriaMedicoGS=gastoMateriaMedicoGS.assign(salud_key=3)
gastoSueldoPersonalMedicoGS=gastoSueldoPersonalMedicoGS.assign(salud_key=4)
```

```python
#VOLVEMOS A REPETIR EL PROCESO DE AGRUPAR LOS DATOS, EN ESTE CASO CON LOS DATOS DE SALUD
```

```python
gastoSalud = pd.DataFrame()
gastoSalud = gastoSalud.append(gastoSaludPorcentajePBI)
gastoSalud = gastoSalud.append(gastoconstruccionInstitucionesMedicoGS)
gastoSalud = gastoSalud.append(gastoMateriaMedicoGS)
gastoSalud = gastoSalud.append(gastoSueldoPersonalMedicoGS)

gastoSalud= gastoSalud.reset_index(drop=True)
gastoSalud.drop_duplicates(inplace=True)

gastoSalud["tiempo_key"] = gastoSalud["tiempo_key"].replace(valor,clave)
gastoSalud["pais_key"] = gastoSalud["pais_key"].replace(p_valor,p_clave)
```

```python
#REALIZAMOS NUESTRO PRIMER JOIN
#repetiremos este proceso con ciertas variaciones que dependen de los datos que vamos a utilizar
```

```python
educacion_salud = pd.merge(gasto_educacion, gastoSalud, how="outer", on=["pais_key", "tiempo_key"])

educacion_salud.head(10)
```

```python
bajoUmbralPobreza = pd.read_csv("./data/pobreza/debajoTresDolares.csv", delimiter=",")
umbralPobreza = pd.DataFrame()
bajoUmbralPobreza['pais_key']= bajoUmbralPobreza['pais_key'].str.strip()

agnos = [1991,1995,2000,2005,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019]
    
for x in agnos:    
    bajoUmbralPobreza_x = bajoUmbralPobreza[['pais_key', str(x)]]
    bajoUmbralPobreza_x['indicadorPobreza'] = bajoUmbralPobreza_x[str(x)]
    bajoUmbralPobreza_x.drop(str(x), axis=1, inplace=True)
    bajoUmbralPobreza_x=bajoUmbralPobreza_x.assign(tiempo_key=x)
    print(bajoUmbralPobreza_x)
    umbralPobreza = umbralPobreza.append(bajoUmbralPobreza_x)
```

```python
accesoAguaPotable = pd.read_csv("./data/pobreza/PorcentajePersonasAccesoAguaPotable.csv", delimiter=",")


agua_index=accesoAguaPotable[accesoAguaPotable['RESIDENCE AREA']!='Total'].index
accesoAguaPotable=accesoAguaPotable.drop(agua_index)
accesoAguaPotable.drop('RESIDENCE AREA', axis=1, inplace=True)
```

```python
#REALIZAMOS NUESTRO TERCER JOIN CON AÑADIDURA DE LOS DATOS DE POBREZA
```
